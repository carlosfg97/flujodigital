```python

CLPedido = Pedido[(Pedido.Pais == 'CL')]
CLContenedor = Contenedor[(Contenedor.Pais == 'CL')]
CLShowroom =  Showroom[(Showroom.Pais == 'CL')]
CLLogs = Logs[(Logs.Pais == 'CL')]

CLPedido = CLPedido.drop(CLPedido[CLPedido['Código Consultora'] == '(not available)'].index)
CLPedido['Código Consultora']= CLPedido['Código Consultora'].astype(str).astype(int)

CLContenedor = CLContenedor.drop(CLContenedor[CLContenedor['Codigo_Consultora'] == '(not available)'].index)
CLContenedor['Codigo_Consultora']= CLContenedor['Codigo_Consultora'].astype(str).astype(int)



CLShowroom = CLShowroom.drop(CLShowroom[CLShowroom['Codigo_Consultora'] == '(not available)'].index)
#CLVenta
```

```python
cont_pedido = CLContenedor.merge(CLPedido, left_on='Codigo_Consultora', right_on='Código Consultora', how='inner')
```
